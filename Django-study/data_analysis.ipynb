```python
# DF (DataFrame = Excel)
data = {
    "country": ["kor", "usa", "china", "japan"],
    "rank": [1,2,3,4],
    "grade": ["A", "B", "C", "D"]
}

df = pd.DataFrame(data) # csv to json, json to csv, xlsx

df
```

```python
# 데이터 셀랙션 => 데이터를 불러오는 방법
# (1) df.컬럼, df['컬럼']

df.grade

df[['country', 'rank', 'grade']]
```

```python
# (2) df.loc[인덱스값, 컬럼명] // df,iloc[] => location기반 항상 숫자값을 필요로 합니다.
df.loc[:, ['grade', 'country']]

# df.loc[df['rank'] > 2, ['grade', 'country']]
df[df['rank'] > 2][['grade', 'country']]
```

```python
# 삭제하는 방법 (drop)
# (1) 단순하게 행 데이터를 삭제하는 방법
df.drop([3], inplace=True)
```

```python
# (2) 컬럼을 삭제
df = df.drop('rank', axis=1)
```

```python
df['rank'] = [1,2,3]
df
```

```python
# 기술통계
df.info()
df.describe()
```

```python
# 데이터 정령 => sort_index or sort_values()
df['new_rank'] = [10, 100, 30]
df

df.sort_values(by='new_rank') # 오름차순 정렬 ascending=True 기본 값
df.sort_values(by='new_rank', ascending=False) # 내림차순
```

# 영화 데이터 분석

```python
import pandas as pd

json_data = {
  "columns": ["Movie", "Release Year", "Audience", "Rating"],
  "index": [0, 1, 2, 3, 4, 5, 6, 7],
  "data": [
    ["Avengers", 2012, 1500, 8.8],
    ["Interstellar", 2014, 1100, 9.1],
    ["Frozen", 2013, 1020, 8.5],
    ["About Time", 2013, 950, 8.7],
    ["The Dark Knight", 2008, 1300, 9.0],
    ["Inception", 2010, 1200, 8.8],
    ["La La Land", 2016, 800, 8.6],
    ["Toy Story", 2010, 980, 8.5]
  ]
}

# JSON 데이터를 DataFrame으로 변환
df = pd.DataFrame(json_data['data'], columns=json_data['columns'])
df
```

```python
### 1) 전체 데이터 중에서 'Moive' 정보만 출력하시오
df.Movie
df['Movie']

### 2) 전체 데이터 중에서 'Movie', 'Rating' 정보를 출력하시오.
df[['Movie', 'Rating']]

### 3) 2013년 이후에 개봉한 영화 데이터 중에서 'Movie', 'Rating' 정보를 출력하시오.
df[df['Release Year'] > 2013][['Movie', 'Rating']]
df.loc[df['Release Year'] > 2013, ['Movie', 'Rating']]  # 둘다 결과는 같다.

### 4) 주어진 계산식을 참고하여 'Recommend' Column을 추가하시오.
### Recommend = (Audience * Rating) // 100
df['Recommend'] = (df['Audience'] * df['Rating']) // 100
df

### 5) 전체 데이터를 'Release Year' 기준 내림차순으로 출력하시오.
df.sort_values(by='Recommend', ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
```

```python
x = np.arange(0,12,1)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y, label="sin", marker="o")
plt.plot(x,y2, label="cos", marker="x")
plt.legend(loc=(1,0.1))

plt.title("sin graph")
plt.xlabel("Time")
plt.ylabel("Amplitube")
plt.grid()

plt.xlim(0, np.pi)
plt.ylim(-1, 1)
```

```python
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```
