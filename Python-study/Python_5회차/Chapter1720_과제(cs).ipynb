[문제3]

클래스 상속에 대해 설명해주세요

클래스의 상속은 부모 클래스로부터 멤버(필드, 메서드, 이너 클래스)를 내려받아 자식 클래스 내부에 포함시키는 문법 요소 입니다. 위 이미지 처럼 공통으로 사용하는 필드와 메서드를 부모 클래스에서 정의하여 자식 클래스에서는 해당 클래스에서만 사용하는 멤버만 정리하여 활용할 수 있습니다.

[문제4]

클래스 다중 상속과 관련된 MRO에 대해 설명해주세요

다중 상속은 슈퍼클래스에 같은 메소드가 정의되어 있을 때, 어떤 메소드를 호출해야 하는지 모호해지는 문제가 있습니다.   
파이썬은 상속 그래프를 특정 순서에 따라서 조회합니다. 이 순서를 Method Resolution Order(MRO), 즉 메서드 결정 순서 라고 합니다. __mro__ 속성은 현재 클래스부터 object 클래스까지 슈퍼 클래스들의 MRO를 튜플 형태로 저장합니다.   
MRO는 서브클래스 정의에 나열된 슈퍼클래스들의 순서도 고려합니다. 이 때, 먼저 선언된 슈퍼클래스의 메소드가 다른 슈퍼클래스들의 메소드를 덮어쓰게 되어, 같은 메소드가 선언되어 있다면 먼저 선언된 슈퍼클래스의 메소드가 호출됩니다.

[문제5]

Self는 어떤 키워드 인가요?

self 는 객체의 인스턴스 그 자체를 말한다.
즉, 객체 자기 자신을 참조하는 매개변수인 셈이다. 객체지향 언어는 모두 이걸 메소드에 안보이게 전달하지만, 파이썬은 클래스의 메소드를 정의할 때 self 를 명시한다. 메소드를 불러올 때 self 는 자동으로 전달된다.

[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

1. TypeError (타입 오류)
무엇인지: 서로 다른 타입의 데이터를 같이 사용할 때 발생해요. 예를 들어, 숫자와 문자열을 더하려고 하면 이 오류가 나요.   
예시: "안녕" + 5 같은 식이죠.
2. KeyError (키 오류)
무엇인지: 딕셔너리에서 없는 키에 접근할 때 발생해요. 딕셔너리는 마치 이름표가 붙어 있는 상자 같은데, 잘못된 이름표를 찾으려 하면 에러가 나요.   
예시: {"이름": "홍길동"}에서 "나이"를 찾으려 하면 발생하죠.   
3. ValueError (값 오류)
무엇인지: 올바른 타입이지만 적절하지 않은 값을 사용할 때 발생해요. 예를 들어, 숫자를 요구하는 곳에 문자 같은 것을 넣으면 에러가 나요.   
예시: "안녕"을 숫자로 바꾸려고 할 때 발생하죠.   
4. AttributeError (속성 오류)
무엇인지: 객체에 없는 속성이나 메소드에 접근하려 할 때 발생해요. 마치 친구에게 없는 전화번호를 물어보는 것과 같아요.   
예시: 강아지 클래스에 "짖기" 기능이 없는데 짖으라고 하면 발생하죠.   
5. FileNotFoundError (파일 없음 오류)
무엇인지: 요청한 파일이 존재하지 않을 때 발생해요. 파일을 열려고 하는데, 그 파일이 없어 찾지 못하는 상황이죠.   
예시: "없는파일.txt"를 열려고 할 때 발생해요.
