[문제1]

list와 tuple의 주된 차이점은 무엇인가요?

튜플과 리스트의 가장 큰 차이는 요솟값을 변화시킬 수 있는지의 여부이다. 즉, 리스트의 요솟값은 변화가 가능하고 튜플의 요솟값은 변화가 불가능하다.

[문제2]

Python의 set에 대해 설명해주세요

파이썬에서 집합(set)이란 중복되는 원소가 없이, 순서에 상관없는 데이터들의 묶음이다. 데이터의 중복을 허용하지 않기 때문에 중복되는 원소가 있다면 하나만을 저장한다. 또한, 순서가 없기 때문에 리스트처럼 인데스 번호를 사용하여 특정 값에 접근할 수는 없다.

[문제3]

Python의 list는 연결리스트(linked list)인가요?

Linked List는 Array List와는 다르게 엘리먼트와 엘리먼트 간의 연결을 이용해서 리스트를 구현한 것을 의미한다. python 의 경우 list 기본 자료형에 linked list 기능이 함께 포함되어 있다.
